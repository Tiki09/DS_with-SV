### 1.	Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
def Num(n):
    if n>0:
        print('{} is +ve'.format(n))
    elif n<0:
        print('{} is -ve'.format(n))
    else:
        print('{} is zero'.format(n))
```

```python
Num(-0.01)
```

### 3. leap year or not

```python
year=int(input('enter any year :'))
if (year%400==0 and year%100==0) or (year%4==0 and year%100!=0):
    print('{}, this is leap year'.format(year))
else:
    print('{}, this is not a leap year'.format(year))
```

```python
def prm(*args):
    for i in args:
        c=False
        if i == 1:
            print(' 1 not prime')
        elif i==2:
            print(' 2 is a prime number')
        else:
            for j in range(2,i):
                if i%j==0:
                    c=True
                    break
            if c==True:
                print('{} is not prime'.format(i))
            else:
                print('{} is prime'.format(i)) 
```
